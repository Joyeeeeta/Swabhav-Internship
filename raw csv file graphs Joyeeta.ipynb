```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the full dataset
df = pd.read_csv("raw_prospect_firmo_data.csv")

# Set aesthetics
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

```python
# 1. Delivery Status count
plt.figure()
sns.countplot(data=df, x="Delivery Status", order=df["Delivery Status"].value_counts().index)
plt.title("Delivery Status Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# 2. Top 10 Companies by Number of Prospects
plt.figure()
top_companies = df["Company Name"].value_counts().head(10)
sns.barplot(x=top_companies.values, y=top_companies.index)
plt.title("Top 10 Companies by Number of Prospects")
```

```python
# 3. Level Distribution (CXO, Dir, etc.)
plt.figure()
sns.countplot(data=df, x="Level", order=df["Level"].value_counts().index)
plt.title("Job Level Distribution")
```

```python

# 4. Top 10 Departments
plt.figure()
top_departments = df["Department"].value_counts().head(10)
sns.barplot(x=top_departments.values, y=top_departments.index)
plt.title("Top 10 Departments")
```

```python
# 5. Industry Distribution (Top 10)
plt.figure()
top_industries = df["Industry"].value_counts().head(10)
sns.barplot(x=top_industries.values, y=top_industries.index)
plt.title("Top 10 Industries")
```

```python
# 6. Sub Industry (Top 10)
plt.figure()
top_sub_industries = df["SUB Industry"].value_counts().head(10)
sns.barplot(x=top_sub_industries.values, y=top_sub_industries.index)
plt.title("Top 10 Sub-Industries")
```

```python
# 7. Revenue Range Distribution
plt.figure()
sns.countplot(data=df, x="Revenue Range", order=df["Revenue Range"].value_counts().index)
plt.title("Revenue Range Distribution")
plt.xticks(rotation=45)
```

```python
# 8. Employee Range Distribution
plt.figure()
sns.countplot(data=df, x="EMP Range", order=df["EMP Range"].value_counts().index)
plt.title("Employee Range Distribution")
plt.xticks(rotation=45)
```

```python

# 9. Top Company States (US)
plt.figure()
top_states = df["Company State"].value_counts().head(10)
sns.barplot(x=top_states.values, y=top_states.index)
plt.title("Top 10 US States by Prospect Count")
```

```python
# 10. Top Cities by Count
plt.figure()
top_cities = df["Company City"].value_counts().head(10)
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title("Top 10 Cities by Prospect Count")
```

```python
# 12. Count of Titles (Top 10)
plt.figure()
top_titles = df["Title"].value_counts().head(10)
sns.barplot(x=top_titles.values, y=top_titles.index)
plt.title("Top 10 Job Titles")
```

```python

# 15. Prospects by Country
plt.figure()
top_countries = df["Company Country"].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Prospect Count")

plt.show()
```
